## Exact Matching + Fuzzy matching with adaptive matching

```python

# Initialize a dictionary to store the drug-sheet mapping with similarity scores
drug_sheet_mapping = {}

# List to store drugs that are not found after exact matching
unmatched_drugs = []

# Perform exact matching first
for drug in consumption_df['item_description']:
    sheets_with_drug = []
    
    # Iterate over each sheet in the warehouse data workbook
    for sheet_name in excel_file.sheet_names:
        # Load the data for the current sheet
        sheet_df = pd.read_excel(excel_file, sheet_name=sheet_name)
        
        # Check if the 'PRODUCT' column exists in the sheet
        if 'PRODUCT' in sheet_df.columns:
            # Perform exact match in the 'PRODUCT' column
            if drug.lower() in sheet_df['PRODUCT'].str.lower().values:
                drug_sheet_mapping[drug] = (sheet_name, 'Exact Match')  # Label as exact match
                break  # Stop further searching after an exact match

    # If no exact match was found, add to unmatched drugs
    if drug not in drug_sheet_mapping:
        unmatched_drugs.append(drug)

# Perform iterative fuzzy matching for remaining unmatched drugs
threshold = 80

while unmatched_drugs and threshold >= 0:
    remaining_unmatched = []
    
    # Iterate over the unmatched drugs
    for drug in unmatched_drugs:
        best_match = None
        best_score = 0
        best_sheet = None

        # Iterate over each sheet for fuzzy matching
        for sheet_name in excel_file.sheet_names:
            # Load the data for the current sheet
            sheet_df = pd.read_excel(excel_file, sheet_name=sheet_name)
            
            if 'PRODUCT' in sheet_df.columns:
                # Perform fuzzy matching on 'PRODUCT' column
                for product in sheet_df['PRODUCT']:
                    if pd.isna(product):
                        continue  # Skip NaN values
                    
                    # Use rapidfuzz's partial_ratio for fuzzy matching
                    similarity = partial_ratio(drug.lower(), str(product).lower())
                    
                    # Update the best match if this is the highest similarity score so far
                    if similarity >= threshold and similarity > best_score:
                        best_match = (sheet_name, similarity)
                        best_score = similarity
                        best_sheet = sheet_name

        # If a match was found, store it; otherwise, add to remaining unmatched
        if best_match:
            drug_sheet_mapping[drug] = best_match  # Store the best sheet and score
        else:
            remaining_unmatched.append(drug)

    # Update unmatched_drugs with those that were still unmatched
    unmatched_drugs = remaining_unmatched
    threshold -= 1  # Decrease the threshold for the next iteration

# Output the results: found matches and drugs still unmatched (if any)
print(f"{'Drug Name':<40} {'Found In Sheets (with Match Type/Score)'}")
print("-" * 80)
for drug, sheet_info in drug_sheet_mapping.items():
    # Exact match or similarity score
    sheet_str = f"{sheet_info[0]} (Match: {sheet_info[1]})"
    print(f"{drug:<40} {sheet_str}")

# Output any drugs that were not found
if unmatched_drugs:
    print("\nDrugs not found in any sheets after fuzzy matching:")
    print("-" * 60)
    for drug in unmatched_drugs:
        print(drug)
```

To further boost the performance of the fuzzy matching process, we can introduce parallel processing using Python’s concurrent.futures module. This allows us to distribute the matching task across multiple CPU cores, which can significantly speed up the execution, especially when dealing with large datasets.

## Steps for Parallel Processing:
Use concurrent.futures:
The ThreadPoolExecutor or ProcessPoolExecutor can be used to parallelize the matching process across multiple sheets.
Parallelize the Fuzzy Matching:
For each drug, we can process all sheets in parallel, checking for matches concurrently.
Return the Best Match:
Once all sheet matching is done in parallel, we return the best match for each drug.

Key Features with Parallel Processing:
Parallel Fuzzy Matching with ThreadPoolExecutor:

We use the ThreadPoolExecutor from concurrent.futures to parallelize the fuzzy matching across all sheets for each drug.
Each sheet is processed concurrently, which significantly speeds up the fuzzy matching process, especially when there are many sheets to check.
Best Match Selection:

Once all threads return their results, we select the best match for each drug by comparing the similarity scores across sheets.
Simulated Annealing and Cooling:

The simulated annealing logic and threshold adjustments remain the same, but now with parallel processing, we expect faster convergence since the sheet matching is done concurrently.
Cost Function:

The cost function balances the time taken and the number of unmatched drugs. This helps guide the simulated annealing process to optimize both performance and accuracy.

Notes:
ThreadPoolExecutor vs. ProcessPoolExecutor: Since reading from an Excel file and performing fuzzy matching isn't CPU-bound (it's more I/O-bound), ThreadPoolExecutor is used. For more CPU-intensive operations, ProcessPoolExecutor might provide better performance.
Performance Gains: The main performance boost comes from matching each drug across multiple sheets in parallel, rather than sequentially.
